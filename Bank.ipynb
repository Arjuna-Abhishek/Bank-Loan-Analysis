```python
x = df.drop(['y','job','marital', 'education','default','contact', 
             'housing', 'loan', 'month', 'poutcome' ], axis=1)
y = df['y']
```

```python
x=pd.get_dummies(x)
y=pd.get_dummies(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
import pandas as pd
import numpy as np
index1 = y_train[y_train.values == 1].index
index0 = y_train[y_train.values == 0].index

if len(index1) > len(index0):
    higher = index1
    lower = index0
else:
    higher = index0
    lower = index1

random.seed(0)
higher = np.random.choice(higher, size=len(lower))
#converting this to an array
lower = np.asarray(lower)
N_I = np.concatenate((lower, higher))
x_train = x_train.loc[N_I]
y_train = y_train.loc[N_I]
```

```python
df = df.select_dtypes(include=['float128'])
```

```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn import preprocessing
import pandas as pd
from statsmodels.base.model import Model
from sklearn.tree import DecisionTreeClassifier
model= tree.DecisionTreeClassifier()
model  = model.fit(x_train,y_train)
tree.plot_tree(model,max_depth=3)
```

```python
from sklearn.metrics import *
predicted_y = model.predict(x_test)
print(predicted_y)
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test,predicted_y)
precision = precision_score(y_test,predicted_y,pos_label='positive',average='micro')
recall = recall_score(y_test,predicted_y,pos_label='positive',average='micro')
f1 = f1_score(y_test,predicted_y,pos_label='positive',average='micro')
print("Accuracy of the decision Tree is",accuracy)
print("Precision of the decision Tree is",precision)
print("Recall of the decision Tree is",recall)
print("F1 of the decision Tree is",f1)
```
